```python
df_data = pd.read_csv ('earthquake_data copy.csv')
limit = 100
df_data = df_data.iloc[0:limit, :]

world_map = folium.Map(zoom_start=100)

incident = folium.map.FeatureGroup()
for lat,lng, in zip(df_data.Latitude,df_data.Longitude):
    incident.add_child(folium.CircleMarker([lat,lng],radius=5,color='yellow',fill=True,fill_color='blue',fill_opacity=0.6))
world_map.add_child(incident)


```
